```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_excel("/content/drive/MyDrive/Bert/Release_Notes_v0_RawData.xlsx")
```

```python
from simpletransformers.classification import ClassificationModel
```

```python
model = ClassificationModel(
    "bert", "./drive/MyDrive/Bert/bert_model",use_cuda=False
)
```

```python
örnek = data.iloc[264]['Release_Note']
print(örnek)
```

```python
tahmin=model.predict([örnek])
if tahmin[0]==[0]:
  print("new")
elif tahmin[0]==[1]:
  print("fixed")
elif tahmin[0]==[2]:
  print("changed")
elif tahmin[0]==[3]:
  print("other")
else:
  print("unresolved")
```
