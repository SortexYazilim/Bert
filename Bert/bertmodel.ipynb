```python
from simpletransformers.classification import ClassificationModel
```

```python
data['labels'] = pd.factorize(data.Label)[0]
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(data, test_size=0.2, random_state=42)
```

```python
import re
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
ps = PorterStemmer()

stop = nltk.download("stopwords")
```

```python

for j in range(800):
    metin = re.sub("[^a-zA-Z]", " ",data["Release_Note"][j])
    metin = metin.lower()
    metin = metin.split()
    metin = [ps.stem(i) for i in metin if not i in set(stopwords.words("english"))]
    son_metin = " ".join(metin)
    data["Release_Note"][j]= son_metin
print(data["Release_Note"])
```

```python
data["Release_Note"][0]
```

```python
train=train[["Release_Note","labels"]]
test=test[["Release_Note","labels"]]
```

```python

train["Release_Note"]=train["Release_Note"].apply(lambda r: str(r))
train['labels']=train['labels'].astype(int)
```

```python
model = ClassificationModel('bert', 'bert-base-uncased',
                            num_labels=5, use_cuda=False,
                            args={'reprocess_input_data': True, 'overwrite_output_dir': True, 
                                  'num_train_epochs': 10, 
                                    "train_batch_size": 64 , "fp16":False, 
                                     "output_dir": "./drive/MyDrive/Bert/bert_model"})
```

```python
model.train_model(train)
```

```python
result, model_outputs, wrong_predictions = model.eval_model(test)
```

```python
predictions = model_outputs.argmax(axis=1)
```

```python
actuals = test.labels.values
```

```python
predictions[:]
```

```python
actuals[:]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(actuals, predictions)
```
